```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df= pd.read_csv('Mall_Customers.csv')
```

```python
df.info()
```

```python
df['Gender'].value_counts()
```

```python
df['Gender']= df['Gender'].str.lower().str.strip()
```

```python
df.head()
```

```python
X=df.drop(['CustomerID'],axis=1)
X['Gender']= X['Gender'].replace({'male':0,'female':1})
```

```python
model= KMeans(n_clusters=5, random_state=42)
model.fit(X)
```

```python
model.labels_
```

```python
X['cluster']= model.labels_
X.info()
```

```python
X.head()
```

```python
X.tail()
```

```python
new_application= [[0, 30, 150, 40],
                  [1, 25, 60, 30],
                  [0, 40, 20, 20],
                  [1, 35, 120, 70]]
result= model.predict(new_application)
result
```

```python
Z=X[['Spending Score (1-100)','Annual Income (k$)']]
model.fit(Z)
```

```python
new_application= [[150, 40],
                  [60, 30],
                  [20, 20],
                  [120, 70]]
result= model.predict(new_application)
result
```

```python
Z['cluster']= model.labels_
```

```python
Z.info()
```

```python
Z.head()
```

```python
Z.tail()
```

```python
plt.scatter(Z['Spending Score (1-100)'], Z['Annual Income (k$)'], c=Z['cluster'])
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Annual Income (k$)')
plt.title('Customer Segmentation')
plt.show()
```
